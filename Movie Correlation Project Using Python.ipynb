```python
# import packages that will be used in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# read in the data
df = pd.read_csv(r'C:\Users\EvertonDavidson\Desktop\Data Projects\Python Correlation\movies.csv')
```

```python
#look at the data

df.head()
```

```python
# Check for missing data using for loop

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Droop rows with missing data

df = df.dropna()
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change data type of 'budget and 'gross' columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
df.head()
```

```python
# Order Data

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot of budget vs gross earnings in millions

plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color" : "red"}, line_kws={"color" : "black"})
```

```python
# Looking at correlation

df.corr()
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Giving companies numeric identifier to look at in matrix

df_companynumber = df


for col_name in df_companynumber.columns:
    if(df_companynumber[col_name].dtype == 'object'):
        df_companynumber[col_name]= df_companynumber[col_name].astype('category')
        df_companynumber[col_name] = df_companynumber[col_name].cat.codes
        
df_companynumber
```

```python
correlation_matrix = df_companynumber.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_companynumber.corr()
```

```python
# looking at correlation in grouping

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
#looking at correlation in sorted pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Looking at high correlated pairs (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
